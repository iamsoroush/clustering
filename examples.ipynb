```python
import sys
sys.path.insert(0, './src')
import numpy as np
import clustering as clt
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

# Generate dummy data

```python
# Generate and plot some dummy data

from sklearn.datasets.samples_generator import make_blobs
from sklearn.preprocessing import StandardScaler

centers = [[1, 1], [-1, -1], [1, -1]]
X, labels_true = make_blobs(n_samples=250, centers=centers, cluster_std=0.25,
                            random_state=0)
X = StandardScaler().fit_transform(X)
unique_labels = set(labels_true)
colors = [plt.cm.Spectral(each)
          for each in np.linspace(0, 1, len(unique_labels))]
for label, color in zip(unique_labels, colors):
    args = np.where(labels_true == label)[0]
    points = X[args]
    plt.scatter(points[:, 0], points[:, 1], c=color)
```

# Rank-Order

```python
# Clustering using RankOrder algorithm

es = clt.RankOrderClustering()
predicted_labels = es.fit_predict(X)
unique = set(predicted_labels)
colors = [plt.cm.Spectral(each)
          for each in np.linspace(0, 1, len(unique))]
for label, color in zip(unique, colors):
    args = np.where(predicted_labels == label)[0]
    points = X[args]
    plt.scatter(points[:, 0], points[:, 1], c=color)
```

# Approximated Rank-Order

```python
# Approximated RankOrder

es = clt.ApproximateRankOrderClustering()
predicted_labels = es.fit_predict(X)
unique = set(predicted_labels)
colors = [plt.cm.Spectral(each)
          for each in np.linspace(0, 1, len(unique))]
for label, color in zip(unique, colors):
    args = np.where(predicted_labels == label)[0]
    points = X[args]
    plt.scatter(points[:, 0], points[:, 1], c=color)
```

# Rank-Order Chinese-Whispers

```python
# ChineseWhispers RankOrder

es = clt.ROCWClustering(metric='euclidean')
predicted_labels = es.fit_predict(X)
unique = set(predicted_labels)
colors = [plt.cm.Spectral(each)
          for each in np.linspace(0, 1, len(unique))]
for label, color in zip(unique, colors):
    args = np.where(predicted_labels == label)[0]
    points = X[args]
    plt.scatter(points[:, 0], points[:, 1], c=color)
```
